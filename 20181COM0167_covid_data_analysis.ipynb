### 5. Bar Graph

#### `5.1. State and Population`

```python
State=data['State']
Population=data['population']
plt.figure(figsize=(11,8))
plt.bar(State,Population)
plt.xlabel("State")
plt.ylabel("Population of state")
plt.title("State and Population")

plt.show()
```

From the above plot The State UP has The heighest population than other states

#### `5.2. State and Total Confirmed`

```python
total__confirmed=data["total__confirmed"]
plt.figure(figsize=(11,8))
plt.bar(State,total__confirmed)
plt.xlabel("State")
plt.ylabel("total__confirmed cases")
plt.title("State and Total Confirmed")

plt.show()
```

From the above plot The State MH has The heighest Total confirmed cases than other states

```python
total__tested=data['total__tested']
plt.figure(figsize=(11,8))
plt.bar(State,total__tested)
plt.xlabel("State")
plt.ylabel("total__tested cases")
plt.title("State and Total tested")

plt.show()
```

#### `5.4. State and Total Recovered`

### 6. Pie Charts

#### `6.1. Top 5 Populated states`

```python
top5p=data.sort_values(by='population',ascending=False).head(5)
print("Top 5 Populated states are")
print(list(top5p['State']))
```

```python
slices=list(top5p["population"])
State=list(top5p["State"])
plt.pie(slices,labels=State,shadow=True,startangle=60, autopct="%0.1f%%",rotatelabels = 135,wedgeprops={'edgecolor':'black'})
plt.show()
```

From the above  pie plot The State UP(34.8%) has The heighest population than other states

#### `6.2. Least 5 covid affected states`

```python
Least5c=data.sort_values(by='total__confirmed',ascending=True).head(5)
print("Least 5 covid affected states")
print(list(Least5c['State']))
```

```python
slices=list(Least5c["total__confirmed"])
State=list(Least5c["State"])
plt.pie(slices,labels=State,shadow=True,startangle=60, autopct="%0.1f%%",rotatelabels = 135,wedgeprops={'edgecolor':'black'})
plt.show()
```

From the above pie plot The State MZ(9.6%) has The least affected than other states

### 7. Histogram

#### `7.1. Total Tested`

```python
total__tested=data['total__tested']
plt.hist(total__tested, bins=10, edgecolor='black',color='lime')
plt.xlabel("total__tested")

plt.title(" Frequency of Total Tested cases")
plt.show()
```

There are many tested cases between range(0,0.8)

#### `7.2. Total Confirmed`

### 8. Line Plot

#### `8.1. State and Population`

```python
State=data['State']
Population=data['population']
plt.figure(figsize=(11,8))
plt.plot(State,Population,color='limegreen',alpha=0.8,marker="*")
plt.xlabel("State")
plt.ylabel("Population of state")
plt.title("State and Population")

plt.show()
```

#### `8.2. State and Total Deceased`

```python
total__tested=data['total__tested']
plt.figure(figsize=(11,8))
plt.plot(State,total__tested,color="orange",alpha=.9)
plt.xlabel("State")
plt.ylabel("total__tested cases")
plt.title("State and Total tested")

plt.show()
```

### 9. Scatter Plot

#### 9.1. `Population and total tested.`

```python
Population=data['population']
total__tested=data['total__tested']
plt.figure(figsize=(8,6))
plt.scatter(Population,total__tested,marker="*",s=60,c="#f01dd7")
plt.xlabel('Population')
plt.ylabel("total__tested cases")
plt.title("Population and Total tested")

plt.show()
```

From the above plot,As the population is less there will be less total tested cases

#### `9.2. Population and total deceased.`

```python
Population=data['population']
total__deceased=data['total__deceased']
plt.figure(figsize=(8,6))
plt.scatter(Population,total__deceased,marker="D",s=70,c="#cc0808")
plt.xlabel('Population')
plt.ylabel("total__deceased cases")
plt.title("Population and Total Deceased")

plt.show()
```

From the above plot,As the population is less there will be less total deceased cases

### 10. Subplots

#### `10.1. create a subplot of 2x2 and plot population vs tested, population vs confirmed, population vs  recovered and population vs deceased in four subplots respectively.` 

```python
Population=data['population']
total__confirmed=data['total__confirmed']
total__deceased=data['total__deceased']
total__tested=data['total__tested']
total__recovered=data['total__recovered']

fig, (ax1,ax2) = plt.subplots(nrows=2, ncols=2, sharex=True)
 
ax1[0].scatter(Population,total__tested,marker="d",c='lime')
ax1[1].scatter(Population,total__confirmed,marker="*",c='red')
ax2[0].scatter(Population,total__deceased,marker="|",c='brown')
ax2[1].scatter(Population,total__recovered,marker= "P",c="#f01dd7")
 
ax1[0].set_xlabel("Population")
ax1[0].set_title("population vs tested")
ax1[0].set_ylabel("tested")

ax1[1].set_xlabel("Population")
ax1[1].set_title(" population vs confirmed")
ax1[1].set_ylabel(" confirmed")

ax2[0].set_xlabel("Population")
ax2[0].set_title("population vs deceased ")
ax2[0].set_ylabel("deceased")

ax2[1].set_xlabel("Population")
ax2[1].set_title("population vs recovered")
ax2[1].set_ylabel("recovered")
 

 
plt.tight_layout()
plt.show()
```
